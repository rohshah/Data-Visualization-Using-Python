```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
#Use the FITBIT Dataset shared for Data Visualization
fit_data=pd.read_csv('/content/FitBit data.csv')
```

```python
fit_data.head()
```

```python
#You are free to make as many visualization as you wish, right from scatter plot to Box plot, Violin plot, Swarm plot and whatever comes to your mind
#scatter plot
fit_data.plot.scatter(x='TotalSteps',y='TotalDistance',c='TrackerDistance')
```

```python
#relational plot
sns.relplot(x='TotalSteps',y='TotalDistance',data=fit_data, hue='VeryActiveDistance')
```

```python
#box plot
sns.boxplot(data=fit_data, x="VeryActiveMinutes", y="FairlyActiveMinutes")
```

```python
#violin plot
sns.violinplot(x=fit_data["TotalDistance"])
```

```python
#swarm plot
sns.swarmplot(data=fit_data, x='TotalDistance')
```

```python
#cat plot
sns.catplot(data=fit_data, x="TotalDistance", kind="boxen")
```

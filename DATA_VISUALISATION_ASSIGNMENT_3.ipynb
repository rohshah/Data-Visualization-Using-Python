```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
#Use the Titanic Dataset which is already available in Seaborn library for Data Visualization
data=sns.load_dataset("titanic")
```

```python
data.head()
```

```python
#You are free to make as many visualization as you wish, right from scatter plot to Box plot, Violin plot, Swarm plot and whatever comes to your mind
#scatter plot
data.plot.scatter(x='age',y='fare')
```

```python
#relational plot
sns.relplot(x='survived',y='age',data=data, hue='sex')
```

```python
#box plot
sns.boxplot(data=data, x="pclass", y="age")
```

```python
#violin plot
sns.violinplot(x=data["pclass"],y=data["fare"])
```

```python
#swarm plot
sns.swarmplot(data=data, x='pclass',y='fare')
```

```python
#Histogram graph     
sns.displot(data['embark_town'])
plt.show()
```

```python
#pie plot
data['sibsp'].value_counts().plot(kind='pie')
```

```python
#bar plot
sns.barplot(data=data, x="pclass", y="age", hue="sex")
```

```python
#boxen plot
sns.boxenplot(data=data, x="age", y="sex")
```

```python
sns.heatmap(data.isna())
```

```python
sns.heatmap(data.corr())
```

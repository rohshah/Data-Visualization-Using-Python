```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as snr
```

```python
#importinf iris data
iris=snr.load_dataset('iris')
```

```python
iris.head()
```

```python
#dropping species coloumn as it is catogorical in nature... dropping is basically used for data cleaning
data=iris.drop(['species'],axis=1) #axis=1 is used for dropping any particular entire coloumn
```

```python
data.head()
```

```python
#line graph
ax=data.plot(figsize=(15,8),title='Iris Dataset')
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
```

```python
#histogram
data['petal_length'].diff().plot(kind='hist',bins=200)
```

```python
#bar graph
data.plot(kind='bar')
```

```python
data.iloc[0:5].plot(kind='bar')
```

```python
data.iloc[10].plot(kind='bar')
```

```python
#pie chart
data.iloc[0].plot.pie(figsize=(10,10),fontsize=15,autopct='%.3f') #here autopact used for showing certain number of deimal points
```

```python
data.iloc[0:3].plot.pie(figsize=(10,10), subplots=True, fontsize=15,autopct='%.2f')
```

```python
#scatter plot
data.plot.scatter(x='sepal_length',y='petal_length',c='c') #we can use scatter plot for at max 3 paramters not more than that
```

```python
data.plot.scatter(x='sepal_length',y='petal_length', c='sepal_width')
```
